### Esercitazione - E09 - Diffusion Models

Scrivere e allenare un diffusion models from scratch è lungo e può richiedere tante risorse computazionali. Inoltre, i diffusion model che hanno ottenuto i risultati migliori sono spesso stati allenati su dataset molto grandi (es. ImageNet) e riprodurli su Colab o con GPU commerciali è complesso.
Per questo, in questa esercitazione, dovrai:

* Navigare su [HuggingFace Models](https://huggingface.co/models?sort=trending) e scegliere un diffusion model per la generazione di immagini che sia stato preallenato su un dataset a tuo piacimento.

* [Facoltativo] Se riesci, cerca un diffusion model che abbia due tecniche di campionamento: un DDPM e un DDIM così da confrontare la velocità di esecuzione.


⚠️ Puoi utilizzare direttamente il codice che trovi su HuggingFace!
