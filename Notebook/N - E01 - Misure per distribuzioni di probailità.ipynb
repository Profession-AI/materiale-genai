```python
# Imports

import numpy as np
import torch
```

```python
# Consideriamo due distribuzioni
p = np.random.uniform(low=1, high=10, size=10)
q = np.random.uniform(low=4, high=14, size=10)
```

```python
# Definizione della funzione della divergenza KL
def kl_divergence(p, q):
	return sum(p * np.log(p/q))
```

```python
# Calcoliamo KL(P || Q)
kl_pq = kl_divergence(p, q)
print('KL(P || Q): %.3f bits' % kl_pq)
# Calcoliamo KL(Q || P)
kl_qp = kl_divergence(q, p)
print('KL(Q || P): %.3f bits' % kl_qp)
```

```python
# Definizione della funzione della divergenza Jensen-Shannon
def js_divergence(p, q):
 m = 0.5 * (p + q)
 return 0.5 * kl_divergence(p, m) + 0.5 * kl_divergence(q, m)
```

```python
# Calcoliamo JS(P || Q)
js_pq = js_divergence(p, q)
print('JS(P || Q): %.3f bits' % js_pq)
# Calcoliamo JS(Q || P)
js_qp = js_divergence(q, p)
print('JS(Q || P): %.3f bits' % js_qp)
```
